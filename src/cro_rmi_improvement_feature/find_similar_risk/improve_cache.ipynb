```python
import os

cache_folder = ".cache/embeddings"
file_name_list = os.listdir(cache_folder)
print(len(file_name_list))
file_name_list[:3]
```

```python
prefix_list = []
for file_name in file_name_list:
    prefix = file_name.split("_")[0]
    if prefix not in prefix_list:
        prefix_list.append(prefix)
prefix_list
```

```python
from collections import Counter

prefix_counter = Counter()
for file_name in file_name_list:
    prefix = file_name.split("_")[0]
    if len(file_name.split("_")) != 2:
        print(file_name)
    prefix_counter[prefix] += 1
prefix_counter
```

```python
for file_name in file_name_list:
    file_path = os.path.join(cache_folder, file_name)
    new_file_name = file_name.split("_")
    if len(new_file_name) != 3:
        # print("file name is not correct",file_name)
        continue
    model_name = None
    old_model = new_file_name[1]
    if old_model == "gemini-exp-03-07":
        model_name = "gemini-embedding-exp-03-07"
    if model_name is None:
        continue
    new_file_name = "_".join([new_file_name[0], model_name, new_file_name[2]])
    # rename file name to new_file_name
    new_file_path = os.path.join(cache_folder, new_file_name)
    os.rename(file_path, new_file_path)
```
