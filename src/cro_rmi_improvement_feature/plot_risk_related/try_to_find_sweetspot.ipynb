```python
# I want to plot 2 graph on same plot
import matplotlib.pyplot as plt
import numpy as np

# x start from 50 to 314
# y is x*(x-1)/2
x = np.arange(50, 315)
print(len(x))
y1 = x * (x - 1) / 2
# y2 start from 60 to 250 with same step as x
y2 = np.linspace(60, 250, len(x))
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```
