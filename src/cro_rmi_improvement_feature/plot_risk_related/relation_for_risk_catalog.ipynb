```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np


def get_top_k_risk_by_embedding(
    embeding_value: list[float],
    embeding_value_list: list[list[float]],
    top_k: int,
    method: str = "cosine",
) -> list[int]:
    """
    Get top k most similar risks based on embedding similarity.

    Args:
        embeding_value (list[float]): Query embedding vector
        embeding_value_list (list[list[float]]): List of embedding vectors to compare against
        top_k (int): Number of top similar risks to return
        method (str): Similarity method, currently only supports "cosine"

    Returns:
        list[int]: Indices of top k most similar risks
    """
    # Convert inputs to numpy arrays
    query_embedding = np.array(embeding_value).reshape(1, -1)
    embedding_matrix = np.array(embeding_value_list)

    if method == "cosine":
        # Calculate cosine similarity between query and all embeddings
        similarities = cosine_similarity(query_embedding, embedding_matrix)[0]
    else:
        raise ValueError(f"Unsupported similarity method: {method}")

    # Get indices of top k similar items (excluding the query itself if it exists in the list)
    top_indices = np.argsort(similarities)[::-1][1 : top_k + 1]

    return top_indices.tolist()
```

```python
import pandas as pd
import pickle
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm
import json

relation_df_dict = {}


def process_row(index, row, company, data_df):
    source_risk_name = row["source_risk"]
    target_risk_name = row["target_risk"]

    source_risk_data = get_risk_data(source_risk_name, company, data_df)
    target_risk_data = get_risk_data(target_risk_name, company, data_df)
    risk_a = source_risk_data.to_dict(orient="records")[0]
    risk_b = target_risk_data.to_dict(orient="records")[0]

    edge_relationship_a_b = classify_edge_relationship(risk_a=risk_a, risk_b=risk_b)
    edge_relationship_b_a = classify_edge_relationship(risk_a=risk_b, risk_b=risk_a)
    edge_relationship = finalize_edge_relationship(
        edge_relationship_a_b,
        edge_relationship_b_a,
        risk_a["risk"],
        risk_b["risk"],
    )

    relationship = edge_relationship["relationship"]
    if relationship == "cause_then_effect":
        direction = [1, 0]
    elif relationship == "effect_then_cause":
        direction = [-1, 0]
    elif relationship == "no_relationship":
        direction = [0, 0]
    elif relationship == "be_a_cause_to_each_other":
        direction = [1, -1]
    else:
        raise ValueError(f"Unknown edge_relationship: {relationship}")
    direction = json.dumps(direction)
    return {
        "index": index,
        "direction": direction,
        "ori_relationship": edge_relationship["ori_relationship"],
        "reason": edge_relationship["reason"],
        "relationship": relationship,
    }


for company, df in result_dict.items():
    source_risk_list = []
    target_risk_list = []
    check_duplicate_list = []

    for index, row in df.iterrows():
        risk_name = row["risk_name"]
        top_k_risk_names = row["top_k_risk_names"]
        for top_k_risk_name in top_k_risk_names:
            if (risk_name, top_k_risk_name) in check_duplicate_list or (
                top_k_risk_name,
                risk_name,
            ) in check_duplicate_list:
                continue
            source_risk_list.append(risk_name)
            target_risk_list.append(top_k_risk_name)
            check_duplicate_list.append((risk_name, top_k_risk_name))
            check_duplicate_list.append((top_k_risk_name, risk_name))

    relation_df = pd.DataFrame(
        {"source_risk": source_risk_list, "target_risk": target_risk_list}
    )

    #### LIMIT TO FIRST 10 ROWS FOR TESTING
    # relation_df = relation_df.head(10)

    with ThreadPoolExecutor(max_workers=8) as executor:
        futures = {
            executor.submit(process_row, index, row, company, data_df): index
            for index, row in relation_df.iterrows()
        }

        for future in tqdm(
            as_completed(futures), total=len(futures), desc=f"Processing {company}"
        ):
            result = future.result()
            # print(result["direction"])
            index = result["index"]
            # relation_df.loc[index, "direction"] = result["direction"]
            # relation_df.loc[index, "ori_relationship"] = result["ori_relationship"]
            # relation_df.loc[index, "reason"] = result["reason"]
            # relation_df.loc[index, "relationship"] = result["relationship"]
            relation_df.at[index, "direction"] = result["direction"]
            relation_df.at[index, "ori_relationship"] = result["ori_relationship"]
            relation_df.at[index, "reason"] = result["reason"]
            relation_df.at[index, "relationship"] = result["relationship"]

    pickle.dump(relation_df, open(f"relation_df_{company}.pkl", "wb"))
    relation_df_dict[company] = relation_df
```
