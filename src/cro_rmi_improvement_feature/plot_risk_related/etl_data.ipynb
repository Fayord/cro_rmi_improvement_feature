```python
import os
import pandas as pd

df_path = "250327_data_from_RMI_from_productionbuild_Q3-2024_label 1.xlsx"
df = pd.read_excel(df_path)
print(df.shape)
df.head()
```

```python
df2 = df.groupby(["company", "risk"], as_index=False).agg(
    {
        "risk_desc": set,
        "iscustom": set,  #'first', # risk
        "company_sector": set,
        "risk_subcat": set,
        "control_name": set,
        "control_des": set,
        "rootcause": set,
        "rootcause_desc": set,
        "process": set,
        "process_desc": set,
    }
)
```

```python
select_columns = [
    "risk",
    "risk_desc",
    "rootcause",
    "rootcause_desc",
    "process",
    "process_desc",
]
df4 = df3[select_columns]
print(df4.shape)
df4.head()
```

```python
# select first row of column risk_desc
cell = df4.iloc[4, 4]
cell = df4.iloc[2, 3]
print(type(cell))
print(cell)
import json
import traceback


def cell_to_list(cell):
    empty_element_list = ["", "-"]
    if type(cell) == set:
        new_cell = []
        for list_string in cell:
            if list_string in empty_element_list:
                continue
            try:
                # if list_string count "'" or '"' have to be even and not 0
                if (
                    (list_string.count("'") % 2 != 0) and (list_string.count("'") != 0)
                ) or (
                    (list_string.count('"') % 2 != 0) and (list_string.count('"') != 0)
                ) is False:
                    new_cell.append(list_string)
                    continue
                print(f"pass list_string :{type(list_string)} {list_string}")

                # swap "'" to '"' and swap '"' to "'"
                list_string = list_string.replace('"', "###")
                list_string = list_string.replace("'", '"')
                list_string = list_string.replace("###", "'")
                list_cell = json.loads(list_string)
                for i in list_cell:
                    if i in empty_element_list:
                        continue
                    new_cell.append(i)
            except:
                print(f"list_string : {list_string}")
                print(traceback.format_exc())
                new_cell.append(list_string)
        return new_cell
    return cell


print(cell_to_list(cell))
print(type(cell_to_list(cell)))
```
